```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


Coffee_Sales= pd.read_csv('Coffee_Sales.csv')
Coffee_Sales=Coffee_Sales.head(50)
print(Coffee_Sales.to_string())
```

```python
sns.histplot(Coffee_Sales['coffee_name'],bins=20)
```

```python
sns.pairplot(Coffee_Sales,hue='coffee_name',palette='Reds',diag_kind='hist')
```

```python
sns.barplot(x='coffee_name',y='money',data=Coffee_Sales)
```

```python
sns.countplot(x='money',data=Coffee_Sales)
```

```python
sns.jointplot(x='coffee_name',y='cash_type',data=Coffee_Sales,kind='scatter')
```
